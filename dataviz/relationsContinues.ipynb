```python
import pandas as pd
import seaborn as sns
import numpy as np 
import matplotlib


data = pd.read_csv('../data/sales_predictions.csv')
data.info()
```

```python
sns.relplot(x="date", y="item_price", kind="line", data=data)
```

```python
sample_df = data[['date', 'item_price']].sample(n=50, random_state=1)
sns.relplot(x="date", y="item_price", kind="line", height=20, aspect=0.6, data=sample_df)
```

```python
sample_df['date'] = pd.to_datetime(sample_df['date']) 
sns.relplot(x="date", y="item_price", kind="line", height=20, aspect=0.6, data=sample_df)
```
