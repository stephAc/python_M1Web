```python
import pandas as pd
import matplotlib
import seaborn as sns
import numpy as np 

data = pd.read_csv('../data/house_pricing.csv')
data.info()
```

```python
sns.relplot(x="LotArea", y="SalePrice", legend="brief", data=data)
```

```python
area_price_df = data[['LotArea', 'SalePrice']]
filter_area_price_df = area_price_df[(area_price_df['LotArea'] < 20000) & (area_price_df['SalePrice'] < 500000)]
sns.relplot(x="LotArea", y="SalePrice", legend="brief", data=filter_area_price_df)
```

```python
sns.relplot(x="LotFrontage", y="LotArea", data=data)
```

```python
frontage_area_df = data[['LotFrontage', 'LotArea']]
filter_frontage_area_df = frontage_area_df[(frontage_area_df['LotFrontage'] < 200) & (frontage_area_df['LotArea'] < 100000)]
sns.relplot(x="LotFrontage", y="LotArea", data=filter_frontage_area_df)
```

```python
sns_frontage_area = sns.scatterplot(x="LotFrontage", y="LotArea", data=data)
matplotlib.pyplot.xlim(0,200)
matplotlib.pyplot.ylim(0,100000)
matplotlib.pyplot.show()
```
